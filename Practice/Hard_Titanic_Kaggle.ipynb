# Results

```
def getAvg(learn:Learner, test_df:DataFrame):
  val = 0
  arr = []
  cat_vars = learn.data.train_dl.x.cat_names
  cont_vars = learn.data.train_dl.x.cont_names
  procs = learn.data.train_dl.x.procs
  dep_var = 'Survived'
  data_test = (TabularList.from_df(test_df, cat_names=cat_vars, 
                                   cont_names=cont_vars, procs=procs, 
                                   processor=learn.data.processor)
        .split_none()
       .label_from_df(cols=dep_var)
       .databunch())
  learn.data.valid_dl = data_test.train_dl
  for x in range(10):
    arr.append(float(learn.validate()[1]))
  for x in range(len(arr)):
    val = val + arr[x]
  return print(f'My Average Accuracy: {val/10}%')
```

```
getAvg(learn, test)
```
